```python
target = df['Type']
df = df.drop('Type',axis=1)
```

```python
from sklearn.model_selection import train_test_split
#20% of the data will be used for testing the model, and the remaining 80% will be used for training.
X_train,X_test,y_train,y_test = train_test_split(df,target,test_size =0.20,random_state=42)
```

```python
sns.pairplot(X_train)
```

```python
# From the plot above, we can see that:

#There are three distinct clusters of data points, corresponding to the three different types of wine.
#The first cluster (red wine) is located in the lower left quadrant of the plot.
#The second cluster (white wine) is located in the upper left quadrant of the plot.
#The third cluster (dessert wine) is located in the upper right quadrant of the plot.
#There is some overlap between the clusters, but overall the data is well-separated.
#This suggests that it should be possible to build a model that can accurately classify the type of wine based on its features.
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```

```python
sns.pairplot(X_train)
```

```python
#The sns.pairplot(X_train) suggests the following:

#There are three distinct clusters of data points, corresponding to the three different types of wine.
#The first cluster (red wine) is located in the lower left quadrant of the plot.
#The second cluster (white wine) is located in the upper left quadrant of the plot.
#The third cluster (dessert wine) is located in the upper right quadrant of the plot.
#There is some overlap between the clusters, but overall the data is well-separated.

#This suggests that it should be possible to build a model that can accurately classify the type of wine based on its features.
```

```python
#Clustering with Original Data:



import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Apply K-means clustering to the original dataset
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df)

# Visualize the clustering results
plt.figure(figsize=(10, 8))
plt.scatter(df['Alcohol'], df['Malic'], c=kmeans.labels_)
plt.xlabel('Alcohol')
plt.ylabel('Malic')
plt.title('K-means Clustering Results')
plt.show()

# Evaluate the clustering performance
silhouette_score = silhouette_score(df, kmeans.labels_)
davies_bouldin_score = davies_bouldin_score(df, kmeans.labels_)

print('Silhouette score:', silhouette_score)
print('Davies-Bouldin score:', davies_bouldin_score)


```

```python
#Silhouette score :The silhouette score is 0.58, which indicates that the clusters are well-separated. This suggests that the K-means clustering algorithm has done a good job of identifying the three distinct clusters of data points.

#Davies-Bouldin score:the Davies-Bouldin score is 0.82, which indicates that the clusters are well-separated. This is consistent with the silhouette score, which also suggests that the K-means clustering algorithm has done a good job of identifying the three distinct clusters of data points.

#K-means suggesting:The K-means clustering algorithm has identified three distinct clusters of data points, which correspond to the three different types of wine. This suggests that it should be possible to build a model that can accurately classify the type of wine based on its features.
```

```python
from sklearn.decomposition import PCA
```

```python
# Reflect on the impact of dimensionality reduction on clustering performance.
#Dimensionality reduction using PCA resulted in a significant improvement in clustering performance.
#The silhouette score and Davies-Bouldin score both improved after dimensionality reduction, indicating that the clusters were more well-separated.
#This is likely because PCA removed some of the redundant and noisy features from the data,
#making it easier for the K-means algorithm to identify the three distinct clusters of data points.

#Overall, dimensionality reduction can be a valuable tool for improving clustering performance.
#By reducing the number of features, dimensionality reduction can make it easier for clustering algorithms to find clusters in the data and can also help to improve the interpretability of the clusters.
```
